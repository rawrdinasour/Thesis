### First Run

```python
run1 = []
```

```python
for word in base:
    for syn in wordnet.synsets(word):
        for l in syn.lemmas():
            run1.append(l.name())
```

```python
run1_set = set(run1)
```

```python
len(run1_set)
```

```python
filename = "lexicon"
outfile = open(filename, "wb")
```

```python
pickle.dump(run1_set, outfile)
outfile.close()
```
